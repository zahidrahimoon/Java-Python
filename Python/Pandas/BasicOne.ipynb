### Starting with pandas

#### Read Data from csv file

```python
df =ps.read_csv("employee_data.csv")
```

#### Get Starting Data By default 5

```python
display(df.head())
```

#### Check Data Set Information

```python
df.info()
```

#### Data statistics

```python
display(df.describe())
```

#### Count Employees by department

```python
display(df["Department"].value_counts())
```

#### Find the average salary per department

```python
display(df.groupby("Department")["Salary"].mean())
```

#### Find employees with experience > 10 years

```python
display(df[df["Experience"] > 10])
```

#### Find the top 5 highest-paid employees

```python
display(df.nlargest(5, "Salary"))
```

#### Check for missing values

```python
display(df.isnull().sum())
```

#### Add a new column "Seniority Level" based on experience

```python
def seniority(exp):
    if exp < 5 : return "Junior"
    elif exp < 10 : return "Mid-level"
    else : return "Senior"
df["seniority Level"] = df["Experience"].apply(seniority)
display(df.head())
```
